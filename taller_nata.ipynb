### Ejercicio 32 — Estadísticas de una lista

Pide al usuario 6 números (uno por uno con `input`) y guárdalos en una lista.  
Sin usar funciones como `max()`, `min()` o `sum()`, calcula e imprime:
- El número más alto
- El número más bajo
- El promedio
- Cuántos son mayores al promedio

````python
nums=[]
for i in range(6):
    nums.append(int(input("Número: ")))

mayor=nums[0]
menor=nums[0]
s=0

for n in nums:
    if n>mayor:
        mayor=n
    if n<menor:
        menor=n
    s+=n

prom=s/len(nums)

cont=0
for n in nums:
    if n>prom:
        cont+=1

print("Mayor:",mayor)
print("Menor:",menor)
print("Promedio:",prom)
print("Mayores al promedio:",cont)
````

### Ejercicio 35 — Sistema de votación

Simula una votación entre tres candidatos: `"Ana"`, `"Luis"` y `"María"`.

1. Pide al usuario cuántas personas van a votar
2. Por cada persona, pide su voto (el nombre del candidato)
3. Si el nombre no coincide con ningún candidato, cuenta el voto como `"nulo"`
4. Al final imprime los votos de cada candidato, los votos nulos y el **ganador**
5. Si hay empate entre los primeros, imprímelo también

````python
votos={"A":0,"B":0}

for i in range(3):
    v=input("Voto A/B: ").upper()
    if v in votos:
        votos[v]+=1

print(votos)
if votos["A"]>votos["B"]:
    print("Gana A")
elif votos["B"]>votos["A"]:
    print("Gana B")
else:
    print("Empate")
````
