```python
nota = int(input("Ingresa una nota entre 0 y 100: "))

if nota < 0 or nota > 100:
    print("Nota inválida.")
elif nota >= 90:
    print("Excelente")
elif nota >= 75:
    print("Bueno")
elif nota >= 60:
    print("Aprobado")
else:
    print("Reprobado")
```

```python
num1 = float(input("Ingresa el primer número: "))
operador = input("Ingresa un operador (+, -, *, /): ")
num2 = float(input("Ingresa el segundo número: "))

if operador == "+":
    print("Resultado:", num1 + num2)
elif operador == "-":
    print("Resultado:", num1 - num2)
elif operador == "*":
    print("Resultado:", num1 * num2)
elif operador == "/":
    if num2 == 0:
        print("No se puede dividir entre cero.")
    else:
        print("Resultado:", num1 / num2)
else:
    print("Operador no reconocido.")
```

```python
ventas = [120, 85, 200, 95, 310, 60, 175]

acumulado = 0

for i in range(len(ventas)):
    venta_dia = ventas[i]
    acumulado += venta_dia
    print(f"Día {i + 1}: ${venta_dia}, Acumulado: ${acumulado}")
```

```python
temps = [22, 35, 18, 41, 29, 38, 17, 33, 26, 40]
dias_calor = 0

for i in range(len(temps)):
    temp = temps[i]
    if temp > 30:
        dias_calor += 1
        print(f"Día {i + 1}: {temp}°C (superó 30°C)")

print(f"Días con más de 30°C: {dias_calor}")

if dias_calor > 3:
    print("Alerta calor")
```

```python
inventario = {
    "manzanas": 50,
    "peras": 12,
    "uvas": 0,
    "mangos": 34,
    "fresas": 7
}

total_unidades = 0

for producto, cantidad in inventario.items():
    total_unidades += cantidad

    if cantidad == 0:
        print(f"{producto}: {cantidad} → Agotado")
    elif 1 <= cantidad <= 10:
        print(f"{producto}: {cantidad} → Stock bajo")
    else:
        print(f"{producto}: {cantidad}")

print(f"\nTotal de unidades en la tienda: {total_unidades}")
```

```python
numero = int(input("Ingresa un número entre 1 y 100: "))

while numero < 1 or numero > 100:
    print("Número fuera de rango, intenta de nuevo.")
    numero = int(input("Ingresa un número entre 1 y 100: "))

print(f"Número aceptado: {numero}")
```

```python
secreto = 42
intentos = 0
adivinado = False

while not adivinado:
    numero = int(input("Adivina el número: "))
    intentos += 1

    if numero == secreto:
        adivinado = True
    elif numero < secreto:
        print("Muy bajo, sube.")
    else:
        print("Muy alto, baja.")

print(f"¡Correcto! Necesitaste {intentos} intentos.")
```

```python
while True:
    pwd = input("Crea una contraseña: ")
    errores = []

    # Regla 1: longitud mínima
    if len(pwd) < 8:
        errores.append("Debe tener al menos 8 caracteres.")

    # Regla 2: que NO sea solo números
    if pwd.isdigit():
        errores.append("No puede ser solo números.")

    if errores:
        print("La contraseña no es válida:")
        for e in errores:
            print(" -", e)
        print("Intenta de nuevo.\n")
    else:
        print("Contraseña registrada correctamente.")
        break
```
