```python
import requests

def buscar_pokemon(nombre):
    url = f"https://pokeapi.co/api/v2/pokemon/{nombre.lower()}"
    r = requests.get(url)
    if r.status_code == 200:
        return r.json()
    else:
        print("Pokemon no encontrado")
        return None
```

## Ejercicio 1
Mostrar tipos, stats y habilidades

```python
data = buscar_pokemon("pikachu")

print("Tipos:")
for t in data["types"]:
    print("-", t["type"]["name"])

print("\nStats:")
for stat in data["stats"]:
    print(stat["stat"]["name"], ":", stat["base_stat"])

print("\nHabilidades:")
for habilidad in data["abilities"]:
    print("-", habilidad["ability"]["name"])
```

```python
lista = ["bulbasaur", "charmander", "squirtle"]

for nombre in lista:
    resultado = buscar_pokemon(nombre)
    if resultado is not None:
        print("Nombre:", resultado["name"])
        print("ID:", resultado["id"])
        print()
```

```python
def mostrar_ficha(pokemon):
    if pokemon is None:
        return

    print("Nombre:", pokemon["name"])
    print("ID:", pokemon["id"])

    print("Tipos:")
    for t in pokemon["types"]:
        print("-", t["type"]["name"])

    print("Stats:")
    for stat in pokemon["stats"]:
        print(stat["stat"]["name"], ":", stat["base_stat"])

    print("Habilidades:")
    for h in pokemon["abilities"]:
        print("-", h["ability"]["name"])

    print("-------------------")
```

## Ejercicio 3
Mostrar fichas

```python
pokemon1 = buscar_pokemon("gengar")
pokemon2 = buscar_pokemon("snorlax")

mostrar_ficha(pokemon1)
mostrar_ficha(pokemon2)
```

### Pregunta teórica
Separar las funciones buscar_pokemon() y mostrar_ficha() permite dividir responsabilidades.
Una función obtiene los datos desde la API y la otra se encarga de mostrarlos.
Esto facilita reutilizar código, mantenerlo y seguir buenas prácticas de programación.


```python
def obtener_stat(pokemon, stat_nombre):
    for stat in pokemon["stats"]:
        if stat["stat"]["name"] == stat_nombre:
            return stat["base_stat"]
```

```python
def comparar_pokemon(p1, p2, stat):
    pokemon1 = buscar_pokemon(p1)
    pokemon2 = buscar_pokemon(p2)

    s1 = obtener_stat(pokemon1, stat)
    s2 = obtener_stat(pokemon2, stat)

    print(p1, stat, ":", s1)
    print(p2, stat, ":", s2)

    if s1 > s2:
        print("Gana", p1)
    elif s2 > s1:
        print("Gana", p2)
    else:
        print("Empate")
```

## Ejercicio 4

```python
comparar_pokemon("snorlax", "machamp", "attack")
comparar_pokemon("pikachu", "raichu", "speed")
comparar_pokemon("blastoise", "charizard", "defense")
```

```python
def pokemon_mas_fuerte(lista, stat):
    mejor = None
    mejor_stat = -1

    for nombre in lista:
        p = buscar_pokemon(nombre)
        valor = obtener_stat(p, stat)

        if valor > mejor_stat:
            mejor_stat = valor
            mejor = nombre

    print("El más fuerte en", stat, "es", mejor)
    return mejor
```
